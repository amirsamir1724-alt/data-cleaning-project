```python
import pandas as pd
df = pd.read_csv("Airbnb_Open_Data.csv")
df
```

```python
df.info()
```

```python
df.columns
```

```python
columns_to_drop = ['reviews per month',
       'review rate number', 'calculated host listings count',
       'availability 365', 'house_rules', 'license', 'id']
columns_to_keep = ['NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',
       'number of reviews', 'last review']
```

```python
df = df[columns_to_keep]
```

```python
df
```

```python
#strip خاصة بازالة المسافات الزائدة من بداية ونهاية اسماء الاعمدة
#lower لتحويل الاحرف كلها الي small
#replace استبدال الفراغات او حرف معين في كل الاعمدة
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df
```

```python
#لمعرفة عدد القيم الفارغة في كل عمود
df.isnull().sum()
#للمعرفة بالنسبة المئوية
df.isnull().mean()*100
```

```python
df['host_name'] = df['host_name'].fillna('unknown')
df['last_review'] = df['last_review'].fillna('no review')
df['host_identity_verified'] = df['host_identity_verified'].fillna('unconfirmed')
```

```python
df.isnull().mean()*100
```

```python
df['price'] = df['price'].str.replace('$','').str.replace(',','')
df['service_fee'] = df['service_fee'].str.replace('$','').str.replace(',','')
# لتحويل الاسعار الي ارقام نستخدم numeric
df['price']=pd.to_numeric(df['price'])
df['service_fee']=pd.to_numeric(df['service_fee'])
```

```python
df.info()
```

```python
df['construction_year'] = df['construction_year'].fillna(df['construction_year'].median())
df['minimum_nights'] = df['minimum_nights'].fillna(df['minimum_nights'].median())
df['price'] = df['price'].fillna(df['price'].median())
df['service_fee'] = df['service_fee'].fillna(df['service_fee'].median())
df['number_of_reviews'] = df['number_of_reviews'].fillna(df['number_of_reviews'].median())
```

```python
df.isnull().mean()*100
```

```python
#لمعرفة القيم الموجودة داخل العمود
df['country'].value_counts()
df['country_code'].value_counts()
```

```python
df
```

```python
df['country'] = df['country'].fillna('United States')
df['country_code'] = df['country_code'].fillna('US')
```

```python
df.isnull().mean()
```

```python
df.select_dtypes(include="object").columns
```

```python
df.info()
```

```python
df['last_review']=pd.to_datetime(df['last_review'], errors='coerce')
```

```python
df['review_year']=df['last_review'].dt.year
df['review_month']=df['last_review'].dt.month
```

```python
df['total_cost']=df['price']+df['service_fee']
```

```python
df=df[['name', 'host_id', 'host_identity_verified', 'host_name', 'neighbourhood_group', 'neighbourhood', 'lat', 'long', 'country', 'country_code', 'instant_bookable', 'cancellation_policy', 'room_type', 'construction_year', 'price', 'service_fee','total_cost', 'minimum_nights', 'number_of_reviews', 'last_review', 'review_year', 'review_month']]
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
for col in df.select_dtypes(include='object').columns:
    print(col, ":", df[col].unique()[:10], "\n")
```

```python
df['neighbourhood_group']=df['neighbourhood_group'].replace({'brookln':'Brooklyn','manhatan':'Manhattan'})
df['neighbourhood_group']=df['neighbourhood_group'].fillna('unknown')
df['instant_bookable']=df['instant_bookable'].fillna(False)
df['cancellation_policy']=df['cancellation_policy'].fillna('unknown')
df['room_type']=df['room_type'].str.lower().str.strip()
```

```python
df['building_age'] = 2025 - df['construction_year']
df['is_luxury'] = df['price'] > 200
df['has_reviews'] = df['number_of_reviews'] > 0
```

```python
new_order = [
    'name',
    'host_id', 'host_identity_verified', 'host_name',
    
    'neighbourhood_group', 'neighbourhood', 'lat', 'long',
    'country', 'country_code',
    
    'room_type',
    'instant_bookable', 'cancellation_policy',
    
    'construction_year', 'building_age',
    
    'price', 'service_fee', 'total_cost', 'is_luxury',
    
    'minimum_nights',
    
    'number_of_reviews', 'has_reviews',
    'last_review', 'review_year', 'review_month'
]
df = df[new_order]
```

```python
from matplotlib import pyplot as plt

plt.hist(df['price'] , bins=10 , color='skyblue' ,edgecolor='black')
plt.title('distribution of prices')
plt.xlabel('price')
plt.ylabel('number of listings')
plt.show()
```

```python
df['room_type'].value_counts().plot(kind='bar', color='orange', edgecolor='black')
plt.title('number of listings per room type')
plt.xlabel('room type')
plt.ylabel('number of listings')
plt.show()
```
